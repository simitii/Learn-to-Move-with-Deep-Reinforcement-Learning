## Introduction

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;This project is based on NIPS (Neural Information Processing Systems) 2017 challenge named Learning to Run by Stanford Neuromuscular Biomechanics Laboratory. The task is developing a controller to move a physiologically-based human model. We decided to use deep reinforcement learning method because the task seemed suitable for it.<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A model of human skeleton given in a simulator (OpenSim) and the controller should decide on activation levels (float between 0 and 1) of 18 leg muscles (9 for each).  The controller takes 41 observations to make these decisions. Finally, the reward is computed as the change in position of the pelvis along the x-axis minus the penalty for the use of ligaments.<br>

## Methods

As the heading of the project tells, **deep reinforcement learning** methods are used. There are a lot of different methods(algorithms and other related components) in the field of deep reinforcement learning. After discovering the environment(OpenSim-RL), it seems wise to use off-policy methods. In contrast to on-policy methods which use only the current data, off-policy methods make use of old data. 

Following **Off-Policy** algorithms are used in the experiments:
- [Deep Deterministic Policy Gradient (DDPG)](https://arxiv.org/abs/1509.02971)
- [Soft Actor Critic (SAC)](https://arxiv.org/abs/1801.01290)
- [Twin Dueling Deep Deterministic Policy Gradient (TD3)](https://arxiv.org/abs/1802.09477)

Each of these algorithms is an instance of **Actor-Critic** which is a class of reinforcement learning algorithms.
Here is a scheme showing how Actor-Critic algorithms work in the essense:
<img src="./notebook_files/actor-critic.png" title="Actor Critic Scheme">

#### Differences between these algorithms
DDPG can be thought of the base/simplest one and the others can be seen as if they add small addition on DDPG.
- SAC uses the idea of maximizing **entropy**(from information theory) of policy in order to improve **exploration capabilities** of the base algorithm.
- TD3 adds **target policy noise** and some other small additions on DDPG in order to reduce **function(policy) approximation error**.

## Feature Engineering
Modifications on observations and actions:
- Velocities of the body parts
- Noise on actions

There is a huge space to do feature engineering for this kind of work. However, we thought that adding only velocities of the parts is enough for the aim of learning to move. Morever, without velocities given observation does not form **markovian state**. Observation being markovian state is one of the main assumptions of the reinforcement learning. Therefore, velocities of the body parts are added on observations.


Morover, we added noise to actions of DDPG&TD3 algorithms for the training time in order to improve exploration.


## System used in Training
Environment = OpenSim-RL

### Computationally Expensive Environment = Slow Environment
The simulation environment is computationally expensive. It makes data collecting process really slow. Moreover, deep reinforcement learning necessitates a lot of iterative steps(steps taken on the environment) to converge. Therefore, a system with multiple environment running in parallel is necessary.

### Some Other Problems Related to the Environment
There are some reported bugs related to the combined usage of python3 and the environment.

### What We Did?
In order to deal with the slowness of the environment, we tried to develop a system based on currently available open source tools/scripts. The main feature of the system is **environment farming** which is multiple environment running in parallel. Moreover, our design separates the environment(simulation) and the learning process with a well-defined communication channel between the two.

#### Steps:
- Developing some scripts to manage OpenSim-RL environment
- Developing **environment farming** feature on an existing deep reinforcement learning framework

```python
import re

a = {}

with open("./notebook_files/sac_3.txt", "r") as f:
    for line in f:
        # process line
        found = re.search(r'(.+(\w|[()]))\s+(-?[0-9]+.?[0-9]*e?E?.?[0-9]*)$',line)
        if found:
            feature_name = found.group(1) # feature_name bosluk ve parantez iceriyor
            feature_value_str = found.group(3)
            feature_value = float(feature_value_str)
            if not feature_name in a:
                a[feature_name] = [feature_value]
            else:
                a[feature_name].append(feature_value)
            #print(feature_name +'=' + str(feature_value))
        if re.search(r'Num Paths',line) and not found:
            print(line)

temp = []
dictlist = []
 
for key, value in a.items():
    temp = [key,value]
    dictlist.append(temp)

TRMean_sac=dictlist[23][1]
PolicyLoss_sac=dictlist[2][1]
```

The best rewards obtained for each model is listed below.

```python
TRMax_ddpg=ddpg1['Test Rewards Max'].tolist()
TRMax_td3=td3['Test Rewards Max'].tolist()
TRMax_sac=dictlist[25][1]
print('Maximum Test Reward for DDPG: ' + str(max(TRMax_ddpg)))
print('Maximum Test Reward for TD3: ' + str(max(TRMax_td3)))
print('Maximum Test Reward for SAC: ' + str(max(TRMax_sac)))
```

The success of model is measured by the mean of the test rewards. The test rewards for our three models are plotted together for comparison purposes. DDPG and TD3 models include observation enhancement and noise.

```python
import matplotlib.pyplot as plt
plt.plot(np.arange(1,10000*len(TRMean_ddpg),10000),TRMean_ddpg)
plt.plot(np.arange(1,20000*len(TRMean_td3),20000),TRMean_td3)
plt.plot(np.arange(1,10000*len(TRMean_sac),10000),TRMean_sac)
plt.xlabel("Training Steps")
plt.ylabel("Test Rewards Mean")
plt.legend(['DDPG','TD3','SAC'])
plt.title('Learning Curve Comparison')
plt.show()
```
